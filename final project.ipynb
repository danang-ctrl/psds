```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
path = '/home/dn/Documents/kuliah/psds/datapsds/'
```

```python
aki = pd.read_csv("/home/dn/Documents/kuliah/psds/datapsds/angkakonsumsiikan.csv")
jpb = pd.read_csv("/home/dn/Documents/kuliah/psds/datapsds/jumlahproduksibenih.csv")
pin = pd.read_csv("/home/dn/Documents/kuliah/psds/datapsds/pembudidayaikannasional.csv")
poi = pd.read_csv("/home/dn/Documents/kuliah/psds/datapsds/produkolahanikan.csv")
pbn = pd.read_csv("/home/dn/Documents/kuliah/psds/datapsds/produksibudidayanasional.csv")
ppin = pd.read_csv("/home/dn/Documents/kuliah/psds/datapsds/produksiperlakuanikannasional.csv")
```

```python
aki.drop(aki.loc[aki['ParamKonsumsiIkan']=='Penyediaan ikan untuk konsumsi (total)'].index,inplace = True)
ak = pd.DataFrame(aki.groupby('ParamKonsumsiIkan')['Nilai'].agg('sum'))
px.line(aki,x= 'Tahun',y = 'Nilai', color = 'ParamKonsumsiIkan')
```

```python
jpb.drop(jpb.loc[jpb['Benih']=='semua ikan'].index, inplace=True)
```

```python
benih = pd.DataFrame(jpb.groupby('Benih').sum())
budidaya = pd.DataFrame(jpb.groupby('Budidaya').sum())
benih = benih.drop(columns = ['Tahun','ID'])
budidaya = budidaya.drop(columns = ['Tahun','ID'])
labels = budidaya.index
label = benih.index
fig = make_subplots(1, 2, specs=[[{'type':'domain'}, {'type':'domain'}]],
                    subplot_titles=['Jenis Budidaya', 'Jenis Benih'])
fig.add_trace(go.Pie(labels=labels, values=budidaya.Jumlah, scalegroup='one',
                     name="Jenis Budidaya"), 1, 1)
fig.add_trace(go.Pie(labels=label, values=benih.Jumlah, scalegroup='one',
                     name="Jenis Benih"), 1, 2)
fig.show()
```

```python
fig = px.bar(jpb, x="Tahun", y="Jumlah", color="Benih", title="Total Produksi Benih Ikan Nasional")
fig.show()
```

```python
fig = px.line(jpb,x = 'Tahun', y = 'Jumlah', color = 'Benih', title = 'Petumbuhan Budidaya Ikan')
fig.show()
```

```python
anot = pd.DataFrame(pin.groupby('Tahun')['Jumlah'].agg('sum'))
an = []
for i in range(len(anot.Jumlah)):
    an.append((abs(anot.Jumlah.iloc[i]-anot.Jumlah.iloc[i-1]))/anot.Jumlah.iloc[i])
    an[0]=0
    an[i] = an[i]*100
    an[i] = round(an[i],2)
an
pinn = pd.DataFrame(pin.groupby('Tahun')['Jumlah'].agg('sum'))
px.bar(pinn, x = pinn.index, y = pinn.Jumlah, text = an)
```
